```python

from sentence_transformers import SentenceTransformer
import sys
import os
import numpy as np
# get current working directory
sys.path.append(os.getcwd())

model = SentenceTransformer('all-MiniLM-L6-v2')

with open("./ouptut-keywords.csv", "r") as f:
    keywords = [line.strip() for line in f]

# sentences are encoded by calling model.encode()
embeddings = model.encode(keywords)

# embeddings in a array of numpy vectors

numpy_embeddings = np.array(embeddings)


print(embeddings)
```

```python
import hdbscan
import numpy as np

# Configure the HDBSCAN clusterer
clusterer = hdbscan.HDBSCAN(min_cluster_size=2)

# Fit the model
clusterer.fit(numpy_embeddings)

# Get cluster labels
cluster_labels = clusterer.labels_

# Analyze the results
num_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
noise_points = list(cluster_labels).count(-1)

print(f"Total number of clusters found: {num_clusters}")
print(f"Number of noise points: {noise_points}\n")

# Group keywords by cluster label
clustered_keywords = {}
for i, label in enumerate(cluster_labels):
    if label not in clustered_keywords:
        clustered_keywords[label] = []
    clustered_keywords[label].append(keywords[i])

# Print the summary report
for label, keyword_list in clustered_keywords.items():
    if label != -1:
        print(f"Cluster {label}:")
        for keyword in keyword_list:
            print(f"- {keyword}")
        print()

# Print noise points
if -1 in clustered_keywords:
    print("Noise Points:")
    for keyword in clustered_keywords[-1]:
        print(f"- {keyword}")
```

```python
# prompt: visualize clusters from previous code

!pip install umap-learn matplotlib

import umap
import matplotlib.pyplot as plt

# Reduce dimensions for visualization
reducer = umap.UMAP(n_neighbors=15, n_components=2, min_dist=0.0, metric='cosine')
embedding_2d = reducer.fit_transform(numpy_embeddings)

# Plot the clusters
plt.figure(figsize=(10, 8))
# Use cluster_labels for coloring. Points with label -1 are noise.
scatter = plt.scatter(embedding_2d[:, 0], embedding_2d[:, 1], c=cluster_labels, cmap='Spectral', s=5)
plt.title('HDBSCAN Clustering of Keywords')
plt.xlabel('UMAP Dimension 1')
plt.ylabel('UMAP Dimension 2')
plt.colorbar(scatter, label='Cluster Label')
plt.show()
```
